```python
import numpy as np
import matplotlib.pyplot as plt
import math


N = 100
x = np.zeros((N,1))
z = np.zeros((N,1))
x[0] = np.random.randn()
    
for j in range(1,N):
    
    x[j] = x[j-1]/2 + 25 * x[j-1] / (1 + x[j-1] * x[j-1]) + 8 * math.cos(1.2*j) + math.sqrt(10)*np.random.randn()
    
    z[j] = x[j] * x[j] / 20 + np.random.randn()
```

```python
R = 1
Q = 10

rms = np.zeros((100,1))

for a in range(100):
    x_n1_n1 = np.random.randn()
    P_n1_n1 = 0.001
    
    x_save = np.zeros((N,1))
    
    x_save[0] = x_n1_n1
    
    for j in range (1,N):
        x_n_n1 = x_n1_n1/2 + 25 * x_n1_n1/(1 + x_n1_n1 * x_n1_n1) + 8 * math.cos(1.2*j)
        
        F = (1/2 + 25*(1 - x_n1_n1*x_n1_n1)/((1 + x_n1_n1*x_n1_n1)*(1 + x_n1_n1*x_n1_n1)))
        
        P_n_n1 = Q + F * P_n1_n1 * F
        
        H = x_n_n1/10
        
        err = z[j] - x_n_n1*x_n_n1/20
        
        den = H * P_n_n1 * H + R
        K = P_n_n1*H/den
        
        x_n_n = x_n_n1 + K*err
        P_n_n = P_n_n1 - K*H*P_n_n1
        
        x_save[j] = x_n_n
        
        x_n1_n1 = x_n_n
        P_n1_n1 = P_n_n
        
    
    err_ekf = np.subtract(x,x_save)
    
    rms[a] = math.sqrt(np.dot(err_ekf.T,err_ekf)/N)
    #print (rms)
    
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4,4))
    #ax[0].plot(th_conv[:,0])
    #ax[0].set_xlim(0,N)
    #ax[0].axhline(y=a[0], color='r')


    ax.plot(x_save)
    ax.set_xlim(0,N)
    #ax[1].axhline(y=a[1], color='r')
    ax.set_title("R = %4.3f, Q = %6.5f I"%(R, Q))
mean_rms = np.sum(rms)/100
print (mean_rms)
```

```python
Ns = 50
NT = 50
rms = np.zeros((100,1))

for a in range(100):
    parts = np.random.randn(Ns, 1)
    
    def particle_likelihood(y0, yh, sig):
        lhood =  (1/(2*np.pi*sig))*np.exp(-0.5*((y0-yh)/sig)**2)
        return lhood

    def resample(w,x):
        c = np.zeros((Ns,1))
        c[0] = w[0]
        for i in range(1,Ns):
            c[i] = c[i-1] + w[i]
        ustart = np.random.uniform(0, 1/float(Ns))
        u = np.zeros((Ns,1))
        i = 0
        previ = 0
        for j in range(Ns):
            u = ustart + (1/float(Ns)) * j
            while (c[i] < u):
                i = i + 1
            x[j] = x[i]
            w[j] = 1/float(Ns)
        return w,x
    
    beta = math.sqrt(10)
    sig = 1
    pweights = np.ones((Ns,1)) / float(Ns)
    
    err_p = np.zeros((N,1))
    
    for n in range(1, N):
        
        for p in range(Ns):
            parts[p] = parts[p]/2 + 25*parts[p]/(1 + parts[p]*parts[p]) + 8*math.cos(1.2*n) + beta*np.random.randn()
            
        sh = np.zeros((Ns, 1))
        lhoods = sh.copy()
        for particle in range(Ns):
            x_pf = parts[particle]
            sh[particle] = x_pf * x_pf/20.0
            lhoods[particle] = particle_likelihood(z[n], sh[particle], sig)
    
    
        pweights = np.multiply(pweights, lhoods)
        weightsum = np.sum(pweights)
        pweights = pweights / weightsum
        
        #pwght = pweights
        
        #index = pweights.argmax()
        
        Neff = 1/(np.dot(pweights.T, pweights))
        #print(Neff)
        if Neff < NT:
            #print('t')
            pweights, parts = resample(pweights,parts)
        
        mean = np.dot(pweights.T,parts)/np.sum(pweights)
        
        #print (parts)
        
        err_p[n] = x[n] - mean
        
    rms[a] = math.sqrt(np.dot(err_p.T,err_p)/float(N - 1))
mean_rms = np.sum(rms)/100
print (mean_rms)
```

```python
bns = []
res = 50
frac = 50.0/res
for i in range (res+1):
    bns.append(i*frac - 25)

zbns = []
zfrac = 50.0/res
for i in range (res+1):
    zbns.append(i*zfrac - 25)


def xn_xn1_func(bns,i):
    N = 5000
    
    xn_xn1 = np.zeros((res,res))
    
    xn_res = np.zeros((N,1))
    
    for k in range(res):
        for j in range(N):
            xn_res[j] = bns[k]/2 + 25 * bns[k] / (1 + bns[k] * bns[k]) + 8 * math.cos(1.2*i) + math.sqrt(10)*np.random.randn()
        xn_xn1[k,:], _ = np.histogram(xn_res, bins = bns)
    
    xn_xn1 = xn_xn1 / N
    return xn_xn1

def zn_xn_func(bns,zbns):
    N = 5000
    
    zn_xn = np.zeros((res,res))
    
    zn_res = np.zeros((N,1))
    
    for k in range(res):
        for j in range(N):
            zn_res[j] = bns[k] * bns[k] / 20 + np.random.randn()
        zn_xn[k,:], _ = np.histogram(zn_res, bins = zbns)
    
    zn_xn = zn_xn / N
    return zn_xn
```
